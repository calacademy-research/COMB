```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
wildlife = pd.read_csv("../data/caples_pts_all.csv")
wildlife_pts = gpd.GeoDataFrame(
    wildlife,
    geometry=gpd.points_from_xy(wildlife.X_U10N83, wildlife.Y_U10N83),
    crs="EPSG:26910",
).to_crs("EPSG:3857")
```

```python
wildlife_pts = wildlife_pts[
    [
        "point",
        "cluster_name",
        "geometry",
    ]
]

wildlife_pts.to_parquet("../outputs/caples_points.parquet")
```

```python
wildlife_pts.plot()
plt.show()
```
